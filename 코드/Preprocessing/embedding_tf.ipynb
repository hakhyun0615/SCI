```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../데이터/merge.csv')
```

```python
def del_bracket(x):
      if '(' in x:
            return x[:x.find('(')]
      return x
```

```python
df['단지명'] = df['단지명'].apply(del_bracket)
df['평수'] = df['전용면적(㎡)'].apply(lambda x : x // 3.3058)//10*10
df['거래금액(만원)'] = df['거래금액(만원)'].apply(lambda x : int(x.replace(',','')))
df['이름'] = df['본번'].apply(lambda x : str(x)) + df['계약년월'].apply(lambda x : str(x)[:4]) + df['단지명']
```

```python
# arr에는 [[평수,가격],[평수,가격],,,]
# vec[평수%10] = 가격  => 1개의 vector

from itertools import combinations

def make_vec(arr):
    vecs = []
    for i in range(len(arr)):
        vec = np.zeros((7))
        try:
            vec[int(arr[i][0]//10)] = arr[i][1]
            vecs.append(vec)
        except:
            continue
    
    return list(combinations(vecs, 2))
```

```python
names = df['이름'].drop_duplicates(keep='first')
embedding_vec = []

i = 0
for name in names:
    print('{}/{}'.format(i,len(names)))
    i+=1
    if i%100 == 0:
        print(i)
    same_apart = df[df['이름'] == name][['평수','거래금액(만원)']].values  # 본번/이름/거래월이 동일함
    embedding_vec.extend(make_vec(same_apart))
```

```python
a = pd.DataFrame(embedding_vec)
a.columns = ['0','1']
```

```python
a.to_excel('../../데이터/더미데이터.xlsx')
```
